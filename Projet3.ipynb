```python
#---------------------------------
# Étape 1 : Initialisation du jeu
#---------------------------------

score = 0 # variable globale, compteur pour accumuler les points

# Tâche : Créer un fichier pour sauvegarder les scores des explorateurs.

import os

def initialiser_jeu():
    nom_fichier = "scores.txt"
    
    # Vérifie si le fichier existe
    if os.path.exists(nom_fichier):
        print("Le fichier existe déjà !")
    else:
        # Si le fichier n'existe pas, on le crée et ajoute l'en-tête
        with open(nom_fichier, "w", encoding="utf-8") as fp:
            fp.write("Nom : Score\n")
        print("Fichier créé avec succès !")
```

```python
# Appel de la fonction

initialiser_jeu()
```

```python
#---------------------------------
# Étape 2 : Explorer les planètes
#---------------------------------
# Tâche : Trouver des objets sur différentes planètes.

planetes = ["Mars", "Jupiter", "Saturne", "Neptune", "Venus"] # créé la liste de planètes

def explorer_planete(planetes): # définir la fonction
    global score # variable globale pour accumuler les points
    planetes_restantes = planetes.copy()  # copie pour ne pas modifier la liste originale
    while planetes_restantes:  # tant qu'il reste des planètes à deviner
        devine = input("Devine le nom d'une planète : ").strip()
        
        if devine in planetes_restantes:
            print("Correct! +20 points!")
            score += 20
            planetes_restantes.remove(devine)  # retire la planète déjà devinée
        else:
            print("Faux, essaye encore!")
    
    print(f"Exploration terminée! Score final : {score} points.")
    return score
```

```python
# Appel de la fonction

explorer_planete(planetes)
```

```python
#--------------------------------------
# Étape 3 : Le code du vaisseau spatial
#--------------------------------------
# Tâche : Décoder un message secret

# créer un dictionnaire code_secret qui associe chaque lettre de l'alphabet à un chiffre
code_secret = {chr(65 + i): str(i + 1) for i in range(26)}

# définir la fonction
def decoder_message(message):
    resultat = "" # prépare chaîne de caractères vide
    for chiffre in message.split():  # on suppose que les chiffres sont séparés par un espace
        for lettre, valeur in code_secret.items(): # parcourt les lettres pour trouver celle qui correspond au chiffre
            if valeur == chiffre: # si le chiffre correspond à la valeur de la lettre
                resultat += lettre # ajoute la lettre décodée au résultat
                break # on sort de la boucle
    return resultat # retourne le message entièrement décodé
```

```python
# Appel de la fonction et ajout des points

score += 10
decoder_message("2 15 14 10 15 21 18") # "BONJOUR"
```

```python
#-------------------------------------
# Étape 4 : Le trou noir (Itérateurs)
#-------------------------------------
# Tâche : Tester différentes combinaisons pour traverser un trou noir

import random

# Générateur qui produit toutes les valeurs entre min_val et max_val inclus.
def combinations(min_val, max_val):
    for val in range(min_val, max_val + 1): # max_val inclus avec +1
        yield val
```

```python
# Définir la fonction qui demande au joueur de deviner la combinaison correcte.

def trou_noir(min_val, max_val):
    global score
    # Choisir une combinaison correcte parmi celles générées avec list()
    correcte = random.choice(list(combinations(min_val, max_val)))
    
    # Demande au joueur de deviner la combinaison
    devine = int(input("Devine la combinaison : "))
    while devine != correcte:
        print("Mauvaise combinaison, essaye encore.")
        devine = int(input("Devine la combinaison : "))

    print("Bravo ! Tu as traversé le trou noir.")
    score += 20
    print("Score accumulé: ", score)
```

```python
# Appel de la fonction

trou_noir(39, 45)
```

```python
#-------------------------------------
# Étape 5 : Les coordonnées stellaires
#-------------------------------------
# Tâche : Gérer des coordonnées stellaires.

def lire_coordonnees(): # définir la fonction
    # Lire et afficher les coordonnées depuis le fichier
    with open("coordonnees.txt", "r", encoding="utf-8") as f:
        for ligne in f:
            print(ligne.strip())

def ajouter_coordonnees(x, y):
    # Ajoute de nouvelles coordonnées au fichier
    with open("coordonnees.txt", "a", encoding="utf-8") as f: # écris dans le fichier sans supprimer avec "a"
        f.write(f"X: {x}, Y: {y}\n") # écris les coordonnées
```

```python
# Appel de la fonction

lire_coordonnees() # si premier appel avant d'ajouter les coordonnées, ne renvoit rien car fichier est vide
```

```python
# Appel de la fonction

ajouter_coordonnees(12, 30) # 12 et 30 remplacent x et y 
score += 5
```

```python
# -------------------------------------------
# Étape 6 : Sauvegarder et calculer le score
# -------------------------------------------
# Tâche : Calculer le score total et sauvegarder les résultats

# Créer un compteur pour totaliser les points : compteur appelé score

# Fonction pour sauvegarder le score
def sauvegarder_score(nom_joueur, score):
    with open("scores.txt", "a", encoding="utf-8") as f:
        f.write(f"{nom_joueur}: {score}\n")
```

```python
# Demande le nom du joueur
nom_joueur = input("Entrez votre nom : ")

# Appel de la fonction avec le nom saisi
sauvegarder_score(nom_joueur, score)

# Affichage du score final
print(f"Score final : {score}")
```
